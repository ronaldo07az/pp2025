```python
num_student = int(input("number of student: "))
student = []
for i in range(num_student):
  student_id = input("Student ID: ")
  student_name = input("Student name: ")
  student_DoB = input("Date of birth: ")
  print(f"Student{i+1}:")
  student.append({"ID": student_id, "Name": student_name, "DoB": student_DoB})

num_course = int(input("number of course: "))
course = []
for i in range(num_course):
  course_id = input("Course ID: ")
  course_name = input("Course name: ")
  course.append({"ID": course_id, "Name": course_name})

mark = {}
for c in course:
  marks[c["ID"]] = []
  print(f"Enter mark for course{c['Name']}: ")
  for s in student:
    m = float(input(f"Enter mark for student {s["ID"]} ({s["Name"]}): "))
    marks[c["ID"]].append({"student_id": s["ID"], "mark": m})
for c in course:
  print(f"Course{c['Name']}({c['ID']}):")
  for ro in marks[c["ID"]]:
    print(f"  {ro['student_id']}: {ro['mark']}")
```
